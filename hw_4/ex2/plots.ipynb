```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams.update({'font.size': 16})

import numpy as np

base_time = list(map(lambda x: int(x), open("base/timing.txt").readlines()))
stream_time = list(map(lambda x: int(x), open("streams/timing.txt").readlines()))

sizes = [
    f"$2^{{{i}}}$" for i in range(18, 25)
]

barWidth = 0.25
br1 = np.arange(len(sizes))
br2 = [x + barWidth for x in br1]

plt.bar(br1, base_time, width=barWidth, label="No streams")
plt.bar(br2, stream_time, width=barWidth, label="Streams")

plt.xticks([r + barWidth / 2 for r in range(len(sizes))], sizes)

plt.xlabel("Input size (elements in vector)")
plt.ylabel("Execution time ($\\mu s$)")
plt.suptitle(f"Timing results from parallelized vector add", fontsize=18)
plt.title(f"Using S_SEG=$2^{{{16}}}$ and 4 streams", fontsize=12)

plt.legend()
```
